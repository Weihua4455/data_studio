```python
import pandas as pd
pd.set_option('display.width', 1000)

from bs4 import BeautifulSoup

import datetime
import time
import calendar

import requests
```

## 1. Get press briefing links from https://obamawhitehouse.archives.gov/briefing-room/statements-and-releases?term_node_tid_depth=40

```python
bf_list = []
```

```python
for i in range(1, 1177):
    url = "https://obamawhitehouse.archives.gov/briefing-room/statements-and-releases?term_node_tid_depth=40&page=" + str(i)
    raw_html = requests.get(url).content
    soup = BeautifulSoup(raw_html, "html.parser")

    print("start page", i)
    
    releases = soup.find("div", attrs={"class": "view-content"}).find_all("div", attrs={"class": "views-row"})
    
    for r in releases:
        bf_dict = {}

        time = r.find("span", attrs={"class": "field-content"}).string
        title = r.find("a").string
        try:
            url = "https://obamawhitehouse.archives.gov/" + r.find("a")['href']
        except:
            url = "unknown"

        bf_dict.update({
                "title" : title,
                "url" : url,
                "time" : time
            })

        bf_list.append(bf_dict)

    print("finished page", i)
```

```python
df = pd.DataFrame(bf_list)
df.head()
```

## Get day of the week from `time`

```python
def get_weekday(row):
    day = datetime.datetime.strptime(row["time"], "%B %d, %Y")
    weekday = calendar.day_name[day.weekday()]
    return weekday
```

```python
df["weekday"] = df.apply(get_weekday, axis=1)
```

## Get text for each url

```python
def get_content(row):
    url = row["url"]
    
    try:
        raw_html = requests.get(url).content
        soup = BeautifulSoup(raw_html, "html.parser")

        print(url)

        content = soup.find("div", attrs={"class": "field-items"}).findChildren()
        content_list = []
        for c in content:
            sentence = c.contents

            content_list.append(sentence)
    except:
        content_list = "not found"
    print("-----------")
    
    return content_list
```

```python
df1["content"] = df1.apply(get_content, axis=1)
```
