```python
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
pd.set_option("display.max_colwidth", 1000)

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("projects_all_somewhat_clean.csv")
words_long = pd.read_csv("word_freq_long.csv")
words_wide = pd.read_csv("word_freq_wide.csv")
```

```python
ax = df.year.value_counts().reset_index().sort_values("index").plot(rot = 0, x = "index", y = "year", kind = "bar", color = "#ddb200", width = 0.8, figsize = (15, 5))
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

ax.tick_params(which = "major", left = False, bottom = False)

ax.set_title("Number of Master's Project Published Every Year", pad = 10)
ax.title.set_fontsize(25)
```

```python
ax = df["title_word_count"].plot(kind = "hist", color = "#a50000", width = 1, figsize = (15, 5), bins = 40)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

ax.tick_params(which = "major", left = False, bottom = False)
ax.set_xticks([0, 5, 10, 15, 20, 25, 30, 35, 40, 50])
```

```python
ax = df.plot(kind = "scatter", x = "year", y = "title_word_count", color = "#a50000", figsize = (10, 10), alpha = 0.25, s=120, lw = 0)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

ax.tick_params(which = "major", left = False, bottom = False)


ax.set_title("The Longest Headline Has 51 Words")
ax.title.set_fontsize(25)
```
